## Création d'un fichier csv contenant le nombre d'images par espèce

```python
rows = len(data)
print(rows)
all_species = data["species"].value_counts()
nb_species = len(all_species)
print(nb_species)
```

### Récupération du dataset avec l'espèce en index

```python
especes = data["species"].unique()
print(especes)
print(len(especes))
print(especes[3])
```

```python
data2 = data.set_index("species")
display(data2)
```

### Mis à jour du nombre d'images

```python
index = 0
dic = {}
for i in range(nb_species):
    nb_images_old = data.at[index, "nb_photos_cl"]
    #print(data.at[index, "species"])
    name_species = str(especes[i])
    result = data2.loc[name_species]
    if (nb_images_old == 1):
        nb_images = 1
    else:
        nb_images = len(result["Unnamed: 0"].value_counts())
    if (nb_images_old != nb_images):
        print(str(especes[i]) + " : " + str(nb_images))
        print(str(especes[i]) + " old : " + str(nb_images_old))
        for i in range(nb_images):
            data.at[index + i, "nb_photos_cl"] = nb_images
    dic[name_species] = nb_images
    index += nb_images
```
